```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sonlar = np.arange(10)
sonlar
```

```python
plt.plot(sonlar)
plt.show()
```

```python
x=np.arange(-5,5,0.1)
y=x*x
plt.plot(x,y)
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.cos(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y, linewidth=2.5, linestyle='', color='#DC7633', marker = "*", alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(x,y,
         linewidth=2.5,
         linestyle='',
         color='#DC7633',
         marker = "*",
         alpha=0.7)

plt.title('Cos(x) funksiya grafigi')
plt.xlabel('pi')
plt.ylabel('Cos(x)')
plt.grid(color='k', alpha=0.5, linestyle='--')

plt.show()
```

```python
plt.figure(figsize=(10,4))
sin=np.sin(x)
cos=np.cos(x)

plt.plot(x,sin,
         linewidth=2.5,
         linestyle='',
         color='#2ECC71',
         marker = "*",
         alpha=0.7)

plt.plot(x,cos,
         linewidth=2.5,
         linestyle='',
         color='#3498DB',
         marker = "*",
         alpha=0.7)

plt.title('Sin(x) va Cos(x) funksiya grafigi')
plt.xlabel('pi')
plt.legend(['sin(x)','cos(x)'], loc=0, fontsize='x-large')
plt.grid(color='#D2B4DE', alpha=0.5, linestyle='--')
# Rasmni saqlab olamiz:
plt.savefig('sin-cos-grafik.png')
plt.show()
```
