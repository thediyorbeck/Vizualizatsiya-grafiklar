```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/flight_delays.csv")
df.head()
```

```python
df.shape
```

```python
plt.figure(figsize=(14,12))
sns.heatmap(data=df.set_index('Month'), cmap='Blues', linewidths=1, square=True)
plt.xlabel('Aviakompaniyalar')
plt.ylabel('Oylar')
plt.title('Aviakompaniyalarning oylar bo\'yicha kechikish vaqti(minutlarda)')
y_labels=['Yanvar','Fevral','Mart','Aprel','May','Iyun','Iyul','Avgust','Sentabr','Oktyabr','Noyabr','Dekabr']
plt.yticks(np.arange(12)+.5, labels=y_labels, rotation=0)
plt.show()
```

```python
import pandas as pd

# Ma'lumotlar tayyorlash
data = {'Name': ['John', 'Alice', 'Bob'],
        'Math': [85, 90, 75],
        'English': [88, 92, 80],
        'History': [75, 95, 85]}

df = pd.DataFrame(data)

# DataFrame'ni keng formatdan uzun formata o'tkazish
df_melted = pd.melt(df, id_vars=['Name'], var_name='Subject', value_name='Score')

# Natijani chop etish
print(df_melted)
```

```python
sns.catplot(data = df_melted, kind='bar', x='Name', y='Score', hue= 'Subject')
plt.show()
```

```python
#JAVOBINGIZNI SHU YERGA YOZING
df=pd.read_csv('https://raw.githubusercontent.com/thediyorbeck/DS_praktikum_datasets/main/Tashkent_average_climate.csv',index_col=0)
df.head()
```

### 2-VAZIFA. `df` barcha ustunlarini chiziqli grafik ko'rinishida chiqaring

```python
#JAVOBINGIZNI SHU YERGA YOZING
plt.figure(figsize=(14,6))
sns.lineplot(data=df)
plt.title("Toshkent shahri yillik ob-havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.show()
```

### 3-VAZIFA. Grafiklar nomini to'g'rilang (`T`, `Tm` va hokazo matnlarni o'zbek tilida chiqaring.)

```python
#JAVOBINGIZNI SHU YERGA YOZING
labels=["Yillik o'rtacha harorat","Yillik o'rtacha maksimal harorat","Yillik o'rtacha minimum harorat","Yillik o'rtacha shamol tezligi",
      "Yomg'irli kunlar soni","Bo'ronli kunlar soni","Tumanli kunlar soni","To'fonli kunlar soni", "Tornadoli kunlar soni", "Do'l yoqqan kunlar soni"]

plt.figure(figsize=(14,6))
sns.lineplot(data=df)
plt.title("Toshkent shahri yillik ob-havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.legend(labels, loc=1)
plt.show()
```

### 4-VAZIFA. Yuqoridagi grafikda turli ko'rsatkichlar aralashib ketdi. Keling ustun turiga qarab alohida grafiklar chizamiz.

#### 4.1. Haroratni ko'rsatuvchi ustunlarni alohida chiziqli grafik qilib chiqaring.

```python
#JAVOBINGIZNI SHU YERGA YOZING
labels=["Yillik o'rtacha harorat","Yillik o'rtacha maksimal harorat","Yillik o'rtacha minimum harorat"]

plt.figure(figsize=(14,6))
sns.lineplot(data=df[['T','TM','Tm']])
plt.title("Toshkent shahri yillik ob-havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.legend(labels, loc=1)
plt.show()
```

#### 4.2. Kunlarni ko'rsatuvchi ustunlarni alohida grafik ko'rinishida chiqaring

```python
#JAVOBINGIZNI SHU YERGA YOZING
labels=["Yomg'irli kunlar soni","Bo'ronli kunlar soni","Tumanli kunlar soni","To'fonli kunlar soni", "Tornadoli kunlar soni", "Do'l yoqqan kunlar soni"]

plt.figure(figsize=(14,6))
sns.lineplot(data=df[['RA','SN','TS','FG','TN','GR']])
plt.title("Toshkent shahri yillik ob-havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.legend(labels, loc=1)
plt.show()
```

### 5-VAZIFA. Yillik o'rtacha shamol tezligini ustunli grafik ko'rinishida chiqaring.

```python
#JAVOBINGIZNI SHU YERGA YOZING
plt.figure(figsize=(16,6))
plt.title("Toshkent shahri uchun ob-havo ko'rsatkichlari")
sns.barplot(x=df.index, y=df['V'])
plt.xlabel("Yillar")
plt.ylabel("O'rtacha shamol tezligi (km/soat)")
plt.xticks(rotation=90)
plt.show()
```

### 6-VAZIFA. Yomg'rili, qorli, bo'ronli, tumanli va do'l yoqqan kunlar sonini heatmap ko'rinishida chiqaring.

```python
#JAVOBINGIZNI SHU YERGA YOZING
plt.figure(figsize=(30,5))
sns.heatmap(data=df[['RA','SN','TS','FG','GR']].T, square=True, linewidth = 1,cmap='YlGnBu')
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
y_labels = ["Yomg'ir","Qor","Bo'ron","Tuman","Do'l"]
plt.yticks(np.arange(5)+0.5,labels = y_labels)
plt.xlabel("Yillar")
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```
